```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

os.makedirs("../data", exist_ok=True)

# Load the dataset
file_path = os.path.abspath("../data/malicious_phish.csv")
df = pd.read_csv(file_path)

# Display basic info
print("Dataset Info:")
print(df.info())

# Display first few rows
print("\nSample Data:")
print(df.head())

# Check class distribution
plt.figure(figsize=(6,4))
sns.countplot(x="type", data=df, palette="viridis", order=df["type"].value_counts().index)
plt.title("Distribution of URL Types")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()

# Check for missing values
print("\nMissing Values:\n", df.isnull().sum())

# URL length analysis
df["url_length"] = df["url"].apply(len)
plt.figure(figsize=(8,5))
sns.histplot(df["url_length"], bins=50, kde=True, color="blue")
plt.title("URL Length Distribution")
plt.xlabel("URL Length")
plt.ylabel("Frequency")
plt.show()
```
